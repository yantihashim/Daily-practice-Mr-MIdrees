```python
# Variance (short manual method)

import numpy as np

class1 = np.array([50,47,51,49,47])
class2 = np.array([5,20,50,80,130])


mean1 = np.mean(class1)
sq_diff1 = [(x - mean1)**2 for x in class1]   
variance1 = sum(sq_diff1) / len(class1)

print("Class 1 Mean:", mean1)
print("Class 1 squared differences:", sq_diff1)
print("Class 1 Variance:", variance1)



mean2 = np.mean(class2)
sq_diff2 = [(x - mean2)**2 for x in class2]
variance2 = sum(sq_diff2) / len(class2)

print("\nClass 2 Mean:", mean2)
print("Class 2 squared differences:", sq_diff2)
print("Class 2 Variance:", variance2)
```

```python
mean2 = sum(class2) / len(class2)
print("\nClass 2 mean is: ", mean2)

sum_of_differences2 = np.array([(i - mean2)**2 for i in class2])
variance2 = (sum_of_differences2.sum()) / len(class2)
print("Class 2 variance is: ", variance2)
```

```python
# variance
import numpy as np

# classes
class1 = np.array([50,47,51,49,47])
class2 = np.array([5,20,50,80,130])

# ---------------- CLASS 1 ----------------
mean1 = sum(class1) / len(class1)
print("Class 1 mean is:", mean1)

sum_of_differences1 = np.array([(i - mean1)**2 for i in class1])
variance1 = sum_of_differences1.sum() / len(class1)
print("Class 1 variance is:", variance1)

# ---------------- CLASS 2 ----------------
mean2 = sum(class2) / len(class2)
print("Class 2 mean is:", mean2)

sum_of_differences2 = np.array([(i - mean2)**2 for i in class2])
variance2 = sum_of_differences2.sum() / len(class2)
print("Class 2 variance is:", variance2)
```
