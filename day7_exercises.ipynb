```python
# create a list of 1 to 10
# use list comprehension
# give me the addition of numbers from 1 to 10

import numpy as np

# list comprehension: [expression for number in range(...)]
nums = [x for x in range(1, 11)]
print("List from 1 to 10:", nums)

# create numpy array from the list (like Idrees' arrayidress)
arrayidress = np.array(nums)
print("NumPy array:", arrayidress)

# addition (sum) of the numbers from 1 to 10
total = arrayidress.sum()   # or np.sum(arrayidress)
print("Addition of numbers from 1 to 10:", total)
```

```python
# Exercise 2: Even numbers from 1 to 20
import numpy as np

evens = [x for x in range(1, 21) if x % 2 == 0]
print("Even numbers from 1 to 20:", evens)

even_array = np.array(evens)
print("NumPy array of evens:", even_array)

print("Sum of all even numbers:", even_array.sum())
```

```python
# Exercise 3: Squares of numbers from 1 to 10

squares = [x*x for x in range(1, 10+1)]
print("Squares:", squares)

square_array = np.array(squares)
print("NumPy array of squares:", square_array)

print("Total of squares:", square_array.sum())
```

```python
# Exercise 4: filter numbers greater than 50

arr = np.array([10, 25, 50, 55, 72, 14, 88, 3])
filtered = arr[arr > 50]

print("Original array:", arr)
print("Numbers > 50:", filtered)
print("Sum of filtered numbers:", filtered.sum())
```
